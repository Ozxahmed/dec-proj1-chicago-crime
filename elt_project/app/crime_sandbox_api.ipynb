```python
import requests
import pandas as pd 

APP_TOKEN = "TuDyOSbRSTPWADojwlVd69hq0"
limit = 10
response = requests.get(f"https://data.cityofchicago.org/resource/x2n5-8w5q.json?"
                        f"$$app_token={APP_TOKEN}"
                        f"&$order=date_of_occurrence asc"
                        f"&$limit={limit}")
print(response)
print(response.status_code)
print(response.headers['Last-Modified'])
print(response.headers)
#response_data = response.json()
print(response.request)

# print the message
crimeData = response.json()
print(crimeData)
assert response.status_code == 200

# create df
df_crimeData = pd.json_normalize(crimeData)

print(len(df_crimeData))
df_crimeData
```

```python
#Transformation to rename columns
df_crimeData = df_crimeData.rename(columns={
    "case_": "caseID",
    "_iucr": "iucr",
    "_primary_decsription": "primary_description",
    "_secondary_description": "secondary_description",
    "_location_description": "location_description",
    "fbi_cd": "fbi_code",
})
print(df_crimeData)
```

```python
#Transformation to change date to more readable format
df_crimeData.dtypes

df_crimeData['date_of_occurrence'] = pd.to_datetime(df_crimeData['date_of_occurrence']) 
df_crimeData['date_only_of_occurrence'] = df_crimeData['date_of_occurrence'].dt.date
df_crimeData['time_of_occurrence'] = df_crimeData['date_of_occurrence'].dt.time

df_crimeData
```

```python
df_crimeData['time']= pd.to_datetime(df_crimeData['date_of_occurrence'])
```
